```python
liste_joueurs=pd.read_excel(f"{tournois}/liste_joueurs.xlsx", engine='openpyxl')
liste_joueurs
```

```python
def liste_joueurs_to_compos(liste_joueurs):
    """
    Transforme une liste de joueurs en une liste de compositions.
    """
    compos_dict = {}
    for index, row in liste_joueurs.iterrows():
        if not row['Team'] in compos_dict:
            compos_dict[row['Team']]={"C1":None, "C2":None, "C3":None, "J1":None, "J2":None, "J3":None, "Serveur1":None, "Serveur2":None, "Serveur3":None}
            compos_dict[row['Team']]["C1"] = row['Classe']
            compos_dict[row['Team']]["J1"] = row['Pseudo_joueur']
            compos_dict[row['Team']]["Serveur1"] = row['Serveur']
        else:
            if compos_dict[row['Team']]["C2"] is None:
                compos_dict[row['Team']]["C2"] = row['Classe']
                compos_dict[row['Team']]["J2"] = row['Pseudo_joueur']
                compos_dict[row['Team']]["Serveur2"] = row['Serveur']
            elif compos_dict[row['Team']]["C3"] is None:
                compos_dict[row['Team']]["C3"] = row['Classe']
                compos_dict[row['Team']]["J3"] = row['Pseudo_joueur']
                compos_dict[row['Team']]["Serveur3"] = row['Serveur']
            else:
                print(f"Erreur: l'équipe {row['Team']} a plus de 3 joueurs.")
    compos = pd.DataFrame.from_dict(compos_dict, orient='index')
    compos.reset_index(inplace=True)
    compos.rename(columns={'index': 'equipe'}, inplace=True)
    return compos

compos=liste_joueurs_to_compos(liste_joueurs)
compos.to_excel(f"{tournois}/compo.xlsx", index=False, engine='openpyxl')
compos
```
